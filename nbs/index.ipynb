## Impact Factor
> - The impact factor (IF) or journal impact factor (JIF) of an academic journal is a scientometric index which reflects the yearly average number of citations to recent articles published in that journal. 
- It is frequently used as a proxy for the relative importance of a journal within its field; journals with higher impact factors are often deemed to be more important than those with lower ones. 
- The impact factor was devised by __Eugene Garfield__, the founder of the Institute for Scientific Information (ISI). 
- Impact factors are calculated yearly _starting from 1975_ for journals listed in the Journal Citation Reports (JCR). 



$$\text{IF}_{y} = {\text{Citations}_{y-1} + \text{Citations}_{y-2} \over \text{Publications}_{y-1} + \text{Publications}_{y-2}}$$


$$\text{IF}_{2017} = {\text{Citations}_{2016} + \text{Citations}_{2015} \over \text{Publications}_{2016} + \text{Publications}_{2015} } = {32389 + 41701 \over 880 + 902 } = 41.577$$

> - The impact factor is used to compare different journals within a certain field. 
- The Web of Science indexes more than 11,500 science and social science journals.
- [source](https://en.wikipedia.org/wiki/Impact_factor)
